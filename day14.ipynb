```python
"""
引例:
输入m,n,计算c(m,n)的值,c(m,n)表示组合数

Version:1.0
Author: vegetabledog
"""

m = int(input("请输入m的值:"))
n = int(input("请输入n的值:"))

# 分别结算m,n,k的值
fm = 1
for i in range(1, m + 1):
    fm *= i

fn = 1
for j in range(1, n + 1):
    fn *= j

fk = 1
for a in range(1, m - n + 1):
    fk *= a

cmn = fm / fn / fk
print(cmn)
```

```python
"""
引入位置参数,关键字参数的概念
"""


def makeJudgement(a: int, b: int, c: int) -> bool:
    """判断三条边是否能构成三角形

    Args:
        a (int): 边a
        b (int): 边b
        c (int): 边c

    Returns:
        bool: true:构成  false:不构成
    """
    return a + b > c and a + c > b and b + c > a


def makeJudgement1(a: int, b: int, c: int, /) -> bool:
    """学习强制位置参数/
        / 前面的将强制位置参数传递
    Args:
        a (int): 边a
        b (int): 边b
        c (int): 边c

    Returns:
        bool:  true:构成  false:不构成
    """
    return a + b > c and a + c > b and b + c > a


# print(makeJudgement1(a=1, b=2, c=3))
# TypeError: makeJudgement1() got some positional-only arguments passed as keyword arguments: 'a, b, c'


def makeJudgement2(*, a: int, b: int, c: int) -> bool:
    """学习命名关键字参数*,
       * 后面的将全部强制关键字参数.

    Args:
        a (int): 边a
        b (int): 边b
        c (int): 边c

    Returns:
        bool:  true:构成  false:不构成
    """
    return a + b > c and a + c > b and b + c > a


# print(makeJudgement2(1, 2, 3))
# TypeError: makeJudgement2() takes 0 positional arguments but 3 were given
```

```python
"""
可变参数的学习
参数中*args接受多个参数,打包成一个元组.如果没有接受到参数,则返回空元组
参数中**kargs接受多少个关键字参数,打包成一个字典,如果没有接受到参数,则返回空字典
"""


def myAdd(*args):
    mySum = 0
    for i in args:
        if type(i) in (int, float):
            mySum += i
    return mySum


print(myAdd(1, 2, 3))  # 6
print(myAdd())  # 0
print(myAdd(1, 2, 3, "helloworld", 1.1))  # 7.1


def myFoo(**kargs):
    for i, j in kargs.items():
        print(f"{i}:{j}")


# name:vegetabledog
# age:20
myFoo(name="vegetabledog", age=20)
```
