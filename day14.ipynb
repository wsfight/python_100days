```python
"""
引例:
输入m,n,计算c(m,n)的值,c(m,n)表示组合数

Version:1.0
Author: vegetabledog
"""

m = int(input("请输入m的值:"))
n = int(input("请输入n的值:"))

# 分别结算m,n,k的值
fm = 1
for i in range(1, m + 1):
    fm *= i

fn = 1
for j in range(1, n + 1):
    fn *= j

fk = 1
for a in range(1, m - n + 1):
    fk *= a

cmn = fm / fn / fk
print(cmn)
```

```python
"""
引入函数概念:
输入m,n,计算c(m,n)的值,c(m,n)表示组合数

version:1.1
author:vegetabledog
"""


def myFactorial(n: int) -> float:
    fn = 1
    for i in range(2, n + 1):
        fn *= i
    return fn
```
